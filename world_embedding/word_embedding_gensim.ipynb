```python
import os
import pandas as pd
import gensim
```

```python
file_path = 'reviews_Cell_Phones_and_Accessories_5.json.gz'

# Download only if the file doesn't exist
if not os.path.exists(file_path):
    print("Downloading dataset...")
    !wget http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Cell_Phones_and_Accessories_5.json.gz
else:
    print("Dataset already exists, skipping download.")
```

```python
df = pd.read_json('reviews_Cell_Phones_and_Accessories_5.json.gz', lines=True)
df.head()
```

```python
df.shape
```

We only need the reviewText column.

```python
df.reviewText[0]
```

```python
# lets test gensim
gensim.utils.simple_preprocess("They look good and stick good! I just don't like the rounded shape because I was always bumping it and Siri kept popping up and it was irritating. I just won't buy a product like this again")
```

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.epochs
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("word2vec_amazon_cell_accessories_reviews_short.model")
```

```python
model.wv.most_similar("bad")
```

```python
model.wv.similarity(w1="cheap", w2="inexpensive")
```

```python
model.wv.similarity(w1="great", w2="nice")
```

```python
model.wv.similarity(w1="great", w2="iphone")
```

## Summary of Word2Vec Project

In this notebook, we implemented a word embedding model using the `gensim` library. Here's a summary of the steps performed:

1.  **Environment Setup**: Installed `gensim` and imported necessary libraries like `pandas` and `os`.
2.  **Data Acquisition**: Downloaded the Amazon Cell Phones and Accessories reviews dataset (JSON format) and loaded it into a Pandas DataFrame.
3.  **Data Preprocessing**:
    *   Examined the raw text data.
    *   Used `gensim.utils.simple_preprocess` to tokenize and clean the review text into a list of words.
4.  **Model Training**:
    *   Initialized a `Word2Vec` model with a context window of 10 and a minimum word count of 2.
    *   Built the vocabulary and trained the model over 5 epochs.
    *   Saved the trained model for future use.
5.  **Model Evaluation**:
    *   Explored word similarities using `most_similar` (e.g., finding synonyms for "bad").
    *   Calculated semantic similarity scores between pairs of words like ("cheap", "inexpensive") and ("great", "nice").
