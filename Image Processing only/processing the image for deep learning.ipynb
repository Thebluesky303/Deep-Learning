# Converting RGB images to Grayscale image


```python
img = cv2.imread(r'C:\Users\User\Desktop\Bikash Limbu\Deep learning\Image Processing only\dog1.jpg')
```

```python
img.shape
```

```python
type(img)
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
#grayscale image
plt.imshow(grayscale_image, cmap='gray')
```

```python
#saving the image
cv2.imwrite('dog1_grayscale.jpg',grayscale_image)
```
